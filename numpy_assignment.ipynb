__Array Creation and Operations__

We will create an array here and use some functions that are provided by numpy...

```python
import numpy as np                      ## importing numpy library here
arr = np.array([23,45,67,87,45,68])        ## creating array here
arr                                     ## printing array here

np.ndim(arr)                            ## to check the dimentions of array
np.shape(arr)                           ## to check size of array in matrices/ no. of rows and columns in 2d array
np.size(arr)                            ## to check the size of array
arr.dtype                               ## check data type of array

## creating and printing simple array
arr1 = np.array([23,45,67,87,45,68])        
arr1                                        
np.ndim(arr1)                               ## check array dimensions  

## creating and printing 2d array
arr2 = np.array([[0,2,3,4,5,6],
                [7,6,5,4,3,2]])
arr2                                    
np.ndim(arr2)                               ## check array dimensions

## creating and printing 3d array
arr3 = np.array([[[0,1,2,3,4,6],
                [2,5,3,6,8,3]],
                [[4,6,9,2,7,8],
                 [2,5,8,4,6,2]]])
arr3
np.ndim(arr3)                               ## check array dimensions

result = arr1 + arr1
result = arr2 - arr2
result = arr2 * arr2
result = arr2 / arr2
result 
```
