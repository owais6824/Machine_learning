<h2>Cleaning Dataset</h2>

we will use **dropna()** method to removes null values from the dataset and is based on specific criteria.
<ul>
<li>axis = "index" this will remove row with any missing value.</li>
<li>axis = "column" this will remove column with any missing value.</li>
There are many more method but we will discuss these only here...
</ul>


```python
df = pd.read_csv('Employee.csv')            # again importing dataset

print("DATA before dropna method \n")       
print(df.shape)                             # print the original shape of dataset
print(df.isnull().sum())                    # print null values in dataset before dropna method 


df.dropna(axis='index', inplace=True)       # this will remove null values row wise 
print("\nDATA after dropna method \n")
print(df.shape)                             # this will print shape after dropna method
print(df.isnull().sum())                    # this will print null values stats.
```

```python
df = pd.read_csv('Employee.csv')

print("\n BEFORE \n")
print(df.shape)
print(df.isnull().sum())

# using axis = column
df.dropna(axis='columns', how='any', inplace=True)

print("\n AFTER \n")
print(df.shape) 
print(df.isnull().sum())
```
